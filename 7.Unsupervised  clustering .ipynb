```python
from scipy.spatial import distance
```

```python
a = (0,3,4,5)
b = (7,6,3,-1)
dst = distance.euclidean(a,b)
print(dst)
```

```python
mink = distance.minkowski(a,b,3)
mink
```

```python
ch = distance.chebyshev(a,b)
ch
```

```python
man = distance.cityblock(a,b)
man
```

```python
# Jaccard similarity 
import numpy as np 
from sklearn.metrics import jaccard_similarity_score          # spatial lib has a jaccard func
a = [2,6,5,0]
b = [1,2,5,7]

out = jaccard_similarity_score(a,b)
out
```
